```python
import torch

import torch.nn as nn

import matplotlib.pyplot as plt
```

```python
# Dummy input: a single-channel image with 1 batch size

x = torch.randn(1,1,28,28)
```

```python
# Define 1, 2, and 3-layer conv stacks (no padding for clarity)

conv1 = nn.Sequential(

    nn.Conv2d(1,1, kernel_size=3)  # 26x26
)

conv2 = nn.Sequential(

    nn.Conv2d(1,1, kernel_size=3), # 26x26

    nn.ReLU(),

    nn.Conv2d(1,1, kernel_size=3) # 24x24

)

conv3 = nn.Sequential(

    nn.Conv2d(1,1, kernel_size=3), # 26x26

    nn.ReLU(),

    nn.Conv2d(1,1, kernel_size=3), # 24x24

    nn.ReLU(),

    nn.Conv2d(1,1, kernel_size=3) # 22x22

)
```

```python
# Forward pass

y1 = conv1(x).squeeze().detach()

y2 = conv2(x).squeeze().detach()

y3 = conv3(x).squeeze().detach()
```

```python
# Plotting the outputs

plt.figure(figsize=(15,4))

for i, (y, title) in enumerate(zip([y1, y2, y3], ['1 Layer','2 Layers','3 Layers'])):

    plt.subplot(1, 3, i +1)

    plt.imshow(y.numpy(), cmap='viridis')

    plt.title(title)

    plt.axis('off')

plt.suptitle('Effect of Stacking Convolutional Layers - ( 3X3 Kernel, No Padding')

plt.show()
```

Size shrinks with more layers (no padding).
Receptive field increases—each pixel output now depends on a larger region of the input.
More layers = more abstract features, like edges → textures → shapes.
