```python
import torch

import matplotlib.pyplot as plt

import seaborn as sns
```

assume 5 customers visit a store per hour 
(Î»=5), and we want to model how often different numbers of customers arrive.

```python
# Define the Poisson distribution (lambda = 5)

lambda_ = 5

poison_distribution = torch.distributions.Poisson(rate=lambda_)
```

```python
# Sample 10000 values from the Poisson distribution

samples = poison_distribution.sample((10000,))
```

```python
# Plot the distribution

sns.histplot(samples.numpy(), bins=15, kde=False, stat='probability')

plt.xlabel('Number of events (Customer per hour)')

plt.ylabel('Probability')

plt.title(f"Poisson Distribution (Î»={lambda_})")

plt.show()
```
