```python
import torch

import torch.nn.functional as F

import matplotlib.pyplot as plt
```

define the functions:
Sigmoid
Tanh
ReLU
Leaky ReLU
ELU
Swish

Then weâ€™ll plot:

Function Output Curve
Gradient Curve (1st derivative)

```python
# Input values

x = torch.linspace(-10,10,1000, requires_grad=True)
```

```python
# Activation Functions

def sigmoid(x): return torch.sigmoid(x)

def tanh(x): return torch.tanh(x)

def relu(x): return F.relu(x)

def leaky_relu(x): return F.leaky_relu(x,negative_slope=0.1)

def elu(x): return F.elu(x, alpha=1.0)

def swish(x): return x * torch.sigmoid(x)
```

```python
# Dictionary of functions

activations = {
    "Sigmoid": sigmoid,
    "Tanh": tanh,
    "ReLU": relu,
    "Leaky ReLU": leaky_relu,
    "ELU": elu,
    "Swish": swish
}
```

```python
# Plot activation curves and gradients

plt.figure(figsize=(15,12))


for i , (name, func) in enumerate(activations.items()):

        # Forward pass

        y = func(x)


        # Backward pass (gradient)

        x_grad = None  # Clear previous gradients

        y.sum().backward(retain_graph = True)

        grad = x.grad.clone()


        # Plot function

        plt.subplot(6,2, 2*i+1)

        plt.plot(x.detach(), y.detach(), label=f'{name} Output')

        plt.title(f'{name} Output')

        plt.grid()

        plt.legend()


        # Plot gradient

        plt.subplot(6,2, 2*i+1)

        plt.plot(x.detach(), grad, color='r', label=f'{name} Gradient')

        plt.title(f'{name} Gradient')

        plt.grid()

        plt.legend()

plt.tight_layout()

plt.show()
```

Activation	            Gradient Behavior
Sigmoid	                Vanishes at both ends
Tanh	                Better than sigmoid, still vanishes
ReLU	                Zero gradient when x < 0
Leaky ReLU	            Fixes ReLUâ€™s dead zone
ELU	Smooth              transition, good for convergence
Swish	                Smooth, retains gradient, modern choice
