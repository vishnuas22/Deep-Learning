```python
import numpy as np

import matplotlib.pyplot as plt

import networkx as nx
```

Step 1: Understanding Neuron Behavior

Each neuron in a neural network:

1️⃣ Takes inputs (features)

2️⃣ Applies weights & biases

3️⃣ Passes through an activation function

4️⃣ Sends output to the next layer

We will visualize how this works using a simple neural network.

```python
# Define the Neuron function

def neuron_output ( x1, x2, w1, w2, bias):

    z = w1 * x1 * w2 * x2 + bias # Linear Combination

    return max(0,z)


# Generate input values

x1_values = np.linspace(-1, 1, 100)

x2_values = np.linspace(-1, 1, 100)


# Compute neuron output

outputs = np.array([[neuron_output(x1, x2, 0.5, -0.3, 0.1)for x1 in x1_values]for x2 in x2_values])


# Plot the neuron output


plt.figure(figsize=(10,6))

plt.contourf(x1_values, x2_values, outputs, levels=50, cmap='coolwarm')

plt.colorbar(label = 'Neauron Output after ReLU')

plt.xlabel('Input x1')

plt.ylabel('Input x2')

plt.title('Visualization of a Single Neuron')

plt.show()
```

What This Does

✔️ Plots how a single neuron processes inputs using ReLU activation.

✔️ Shows how neuron output changes for different inputs.


 Step 3: Visualizing a Small Neural Network


visualizing a simple feedforward neural network.

```python
# Create graph

G = nx.DiGraph()


# Define layers

input_nodes = ["x1", "x2"]

hidden_nodes = ["h1", "h2", "h3"]

output_nodes = ["y"]
```

```python
# Add nodes

G.add_nodes_from(input_nodes+ hidden_nodes+ output_nodes)
```

```python
# Add edges (connections)

edges = [

    ('x1','h1'),('x1','h2'),('x1','h3'),
    ('x2','h1'),('x2','h2'),('x2','h3'),
    ('h1','y'),('h2','y'),('h3','y')

]

G.add_edges_from(edges)
```

```python
# Draw the network


plt.figure(figsize=(8,6))

pos = {

    'x1' : (0,2) , 'x2': (0,1),

    'h1' : (1, 2.5), 'h2' : (1, 1.5), 'h3' : (1, 0.5),

    'y':(2, 1.5)
}

nx.draw(G, pos, 
        with_labels = True, 
        node_size =2000, 
        node_color = 'lightblue',
        edge_color = 'grey',
        font_size = 10,
        font_weight = 'bold'
        )

plt.title("Simple Neural Network Visualization")

plt.show()
```

What This Does

✔️ Draws a 3-layer neural network

✔️ Shows how neurons are connected
