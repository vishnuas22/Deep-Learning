```python
import numpy as np

def greet():
    print("\nüëã Hi, Ghost!")

def farewell():
    print("\nüëã Goodbye! Have a great day!")

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return max(0, x)

def leaky_relu(x):
    return max(0.01 * x, x)

# Main Function
def main():

    while True:
        print("\nüîπ Choose an option:")
        print("1. Greet")
        print("2. Sigmoid")
        print("3. Tanh")
        print("4. ReLU")
        print("5. Leaky ReLU")
        print("6. Exit")

        choice = input("\nEnter your choice (1-6): ")

        if choice == "1":
            greet()
        
        elif choice in ["2", "3", "4", "5"]:
            try:
                x = float(input("\nEnter a number for activation function: "))

                if choice == "2":
                    print(f"Sigmoid({x}) = {round(sigmoid(x), 3)}")

                elif choice == "3":
                    print(f"Tanh({x}) = {round(tanh(x), 3)}")

                elif choice == "4":
                    print(f"ReLU({x}) = {relu(x)}")

                elif choice == "5":
                    print(f"Leaky ReLU({x}) = {leaky_relu(x)}")

            except ValueError:
                print("‚ùå Invalid input. Please enter a numeric value.")
        
        elif choice == "6":
            farewell()
            break  # Exits the loop
        
        else:
            print("‚ùå Invalid choice. Try again.")

main()
```
