```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Define Parameters

epochs = 50

initial_lr = 0.1

gammas = [0.99, 0.95, 0.9, 0.85] # Different decay rates
```

```python
plt.figure(figsize=(8,6))

# Plot decay curves for different gamma values

for gamma in gammas:

    lrs = [initial_lr * (gamma ** epoch) for epoch in range(epochs)]

    plt.plot(range(epochs), lrs, label=(f'Gamma: {gamma}'))


# Formatting the plot

plt.xlabel('Epochs')

plt.ylabel('Learning rate')

plt.title('Exponential Decay learning rate for different gammas')

plt.legend()

plt.grid()

plt.show()
```

📌 Expected Graph

Higher gamma (0.99) → Learning rate stays higher for longer.

Lower gamma (0.85) → Learning rate drops much faster.

0.95, 0.9 → Midway balance between fast decay and slow decay.
