```python
import torch

import torch.nn as nn

import torchvision.transforms as transforms

import matplotlib.pyplot as plt

from PIL import Image
```

```python
# Load sample grayscale image from local

# Load the local image
img = Image.open("Husky Love.jpg").convert("L")  # Convert to grayscale
```

```python
transform = transforms.Compose([

    transforms.Resize((128,128)),

    transforms.ToTensor()
])


img_tensor = transform(img).unsqueeze(0) # [1, 1, H, W]
```

```python
# Create a conv layer with 6 filters, kernel size 3x3

conv = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=3, padding=1)

output = conv(img_tensor)
```

```python
# Visualize original + 6 filters

fig, axs = plt.subplots(1,7, figsize=(20,5))

axs[0].imshow(img_tensor.squeeze(), cmap='gray')

axs[0].set_title('Original')

axs[0].axis('off')


for i in range(6):

    axs[i+1].imshow(output[0,i].detach().numpy(), cmap='gray')

    axs[i+1].set_title(f'Filter {i+ 1}')

    axs[i+1].axis('off')  

plt.tight_layout()

plt.show()
```
