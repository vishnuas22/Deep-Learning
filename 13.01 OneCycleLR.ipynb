```python
import torch

import torch.nn as nn

import torch.optim as optim

import matplotlib.pyplot as plt

from torch.optim.lr_scheduler import OneCycleLR
```

```python
# Dummy model (simple linear layer)

model = torch.nn.Linear(10,1)
```

```python
# Optimizer

optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
# OneCycleLR Scheduler

scheduler = OneCycleLR(optimizer,
                       max_lr=0.1,
                       epochs=50,
                       steps_per_epoch=1)
```

```python
# Tracking Learning Rates

lrs =[]

# Simulate 50 training steps

for epoch in range(50):

    optimizer.step()  # Dummy step (just for simulation)

    lrs.append(optimizer.param_groups[0]['lr'])  # Track LR

    scheduler.step() # Update LR


# Plot Learning Rate Schedule

plt.plot(lrs)

plt.xlabel('Epochs')

plt.ylabel('Learning rate')

plt.title("OneCycleLR Learning Rate Schedule")

plt.show()
```
