```python
import numpy as np
import matplotlib.pyplot as plt

# Define Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Generate values for x-axis
x = np.linspace(-5, 5, 400)

# Compute function outputs
sigmoid_values = sigmoid(x)
tanh_values = tanh(x)
relu_values = relu(x)
leaky_relu_values = leaky_relu(x)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, sigmoid_values, label="Sigmoid", color="blue")
plt.title("Sigmoid Activation Function")
plt.axhline(y=0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(x=0, color='black', linewidth=0.5, linestyle="--")
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, tanh_values, label="Tanh", color="red")
plt.title("Tanh Activation Function")
plt.axhline(y=0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(x=0, color='black', linewidth=0.5, linestyle="--")
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, relu_values, label="ReLU", color="green")
plt.title("ReLU Activation Function")
plt.axhline(y=0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(x=0, color='black', linewidth=0.5, linestyle="--")
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, leaky_relu_values, label="Leaky ReLU", color="purple")
plt.title("Leaky ReLU Activation Function")
plt.axhline(y=0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(x=0, color='black', linewidth=0.5, linestyle="--")
plt.grid()

plt.tight_layout()
plt.show()
```
