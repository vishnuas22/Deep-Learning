```python
flat_data_arr = [] # used for collecting input data

target_arr = [] # used for collecting output data

categories = ['Cat','Dog']

datadir = '/content/drive/MyDrive/Data'
```

```python
for i in categories: # cat

   print('Loading Categories...........')

   path = os.path.join(datadir,i) #'/content/drive/MyDrive/Data',cat

   for img in os.listdir(path):  #for j in os.listdir(path)

      img_arr = imread(os.path.join(path,img)) # imread(os.path.join(path,img))

      img_resize = resize(img_arr,(150,150,3)) #height,width,RGB (PIXELS)

      flat_data_arr.append(img_resize.flatten()) # 1D DATA USING FLATTEN

      target_arr.append(categories.index(i)) # 0 for cat and 1 for dog

   print('Loaded completely',i)
```

```python
flat_data = np.array(flat_data_arr)

target_data = np.array(target_arr)

jarvis = pd.DataFrame(flat_data)

jarvis
```

```python
jarvis['Target']=target_arr

jarvis
```

```python
x = jarvis.iloc[:,:-1].values

x
```

```python
y = jarvis.iloc[:,-1]

y
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.20,random_state=42)

train_x
```

```python
from sklearn.svm import SVC

hydra = SVC()

hydra.fit(train_x,train_y)

y_pred = hydra.predict(test_x)

y_pred
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy_score:',accuracy_score(test_y,y_pred))
```
