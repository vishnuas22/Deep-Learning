ðŸ”¹ 1. Bernoulli Distribution (Yes/No Events)

ðŸ‘‰ The simplest probability distributionâ€”only two possible outcomes:

Success (1) with probability p

Failure (0) with probability 1 - p

Example: Tossing a Coin

Probability of Heads = p = 0.5

Probability of Tails = 1 - p = 0.5


ðŸ“Œ Bernoulli Formula:

P(X=x) = p^x(1-p)^(1-x)

where x = 0 or 1, and p is the probability of success.

Deep Learning Usage:

Used in Binary Classification (Sigmoid Activation)

Dropout Regularization (randomly dropping neurons follows Bernoulli)

```python
import torch

import matplotlib.pyplot as plt
```

```python
# Probability of success (e.g., p = 0.7 means 70% chance of success)

p = 0.7

bernoulli_dist = torch.distributions.Bernoulli(torch.tensor([p]))
```

```python
# Sample 1000 values from the Bernoulli distribution

samples = bernoulli_dist.sample((1000,))
```

```python
# Plot the distribution

plt.hist(samples.numpy(), bins=2, edgecolor='black', density=True)

plt.xticks([0,1], ['Failure (0)', 'Success (1)'])

plt.xlabel('Outcome')

plt.ylabel('Probability')

plt.title(f'Bernoulli distribution (p = {p})')

plt.show()
```
